```python
from cvlib.object_detection import YOLO

yolo = YOLO('../model/yolov4-tiny-custom_best.weights', '../yolov4-tiny-test.cfg', '../obj.names')
img = cv2.imread('../obj/stickers1_4_jpg.rf.2adbd1a69334892d08683c34e6a1ce75.jpg')
bbox, label, conf = yolo.detect_objects(img, confidence=0.5, nms_thresh=0.4)
yolo.draw_bbox(img, bbox, label, conf, write_conf=True)

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
with open('../train.txt', 'r') as f:
    dataset = [line.strip() for line in f.readlines()]

total_t = 0

for i in range(len(dataset)):
    t0 = time.time()
    img = cv2.imread(dataset[i])
    bbox, label, conf = yolo.detect_objects(img)
    yolo.draw_bbox(img, bbox, label, conf, write_conf=True)
    t = time.time()
    diff = t-t0
    total_t += (diff)

print('average inference time =  {:.4f}'.format(total_t/len(dataset)))
```

cv2

```python
def init_yolo(config, weights, labels):
    with open(labels, 'r') as f:
        classes = [line.strip() for line in f.readlines()]
    colours = np.random.randint(0, 255, size=(len(classes), 3), dtype='uint8')

    net = cv2.dnn.readNet(config, weights)
    net.setPreferableBackend(cv2.dnn.DNN_BACKEND_OPENCV)

    ln = net.getLayerNames()
    ln = [ln[i - 1] for i in net.getUnconnectedOutLayers()]

    return net, ln, classes, colours

def detect(img, net, ln, classes, colours, score_thresh=0.5, iou_thresh=0.4):
    blob = cv2.dnn.blobFromImage(img, 1/255.0, (416, 416), swapRB=True, crop=False)
    net.setInput(blob)
    outputs = net.forward(ln)

    boxes = []
    confidences = []
    classIDs = []
    h, w = img.shape[:2]

    for output in outputs:
        for detection in output:
            scores = detection[5:]
            classID = np.argmax(scores)
            confidence = scores[classID]
            if confidence > score_thresh:
                box = detection[:4] * np.array([w, h, w, h])
                (centerX, centerY, width, height) = box.astype("int")
                x = int(centerX - (width / 2))
                y = int(centerY - (height / 2))
                box = [x, y, int(width), int(height)]
                boxes.append(box)
                confidences.append(float(confidence))
                classIDs.append(classID)

    indices = cv2.dnn.NMSBoxes(boxes, confidences, score_thresh, iou_thresh)
    if len(indices) > 0:
        for i in indices.flatten():
            (x, y) = (boxes[i][0], boxes[i][1])
            (w, h) = (boxes[i][2], boxes[i][3])
            color = [int(c) for c in colours[classIDs[i]]]
            cv2.rectangle(img, (x, y), (x + w, y + h), color, 2)
            text = "{}: {:.4f}".format(classes[classIDs[i]], confidences[i])
            cv2.putText(img, text, (x, y - 5), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 1)
    
    return img
```

```python
config = '../yolov4-tiny-test.cfg'
weights ='../model/yolov4-tiny-custom_best.weights'
labels = '../obj.names'
net, ln, classes, colours = init_yolo(config, weights, labels)

img = cv2.imread('../obj/nightstickers_2_jpg.rf.9830624cde375da317a81a406bd5120d.jpg')
detect(img, net, ln, classes, colours)

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
with open('../train.txt', 'r') as f:
    dataset = [line.strip() for line in f.readlines()]

total_t = 0

for i in range(len(dataset)):
    t0 = time.time()
    img = cv2.imread(dataset[i])
    img = detect(img, net, ln, classes, colours)
    t = time.time()
    diff = t-t0
    total_t += (diff)

print('average inference time =  {:.4f}'.format(total_t/len(dataset)))
```

#### FPS comparison

cvlib

```python
image_widget = widgets.Image(format='jpeg', width=680, height=460)
display(image_widget)

cap = cv2.VideoCapture('../IMG_4982.mov')
yolo = YOLO('../model/yolov4-tiny-custom_best.weights', '../yolov4-tiny-test.cfg', '../obj.names')

start_time = time.time()
fps_total = 0
frame_total = 0

while (cap.isOpened()):
    ret, frame = cap.read()
    if not ret:
        break

    frame = cv2.resize(frame,(680, 460))
    bbox, label, conf = yolo.detect_objects(frame)
    yolo.draw_bbox(frame, bbox, label, conf, write_conf=True)
    
    cur_time = time.time()
    elapsed_time = cur_time - start_time
    fps = 1/elapsed_time
    fps_total += fps
    frame_total +=1
    start_time = cur_time
    fps_txt = "FPS: {:.2f}".format(fps)
    cv2.putText(frame, fps_txt, (30,430), cv2.FONT_HERSHEY_COMPLEX_SMALL, 1, (0, 0, 255), 1)

    image_widget.value = cv2.imencode('.jpg', frame)[1].tobytes()
    

cap.release()
avg_cvlib = fps_total / frame_total

print("Avg fps: {:.2f}".format(avg_cvlib))
```
